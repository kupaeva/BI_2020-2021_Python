```python
connection = sqlite3.connect('snp.db')
```

```python
connection.execute('''CREATE TABLE genstudio (
SNP_id INTEGER PRIMARY KEY,
SNP_Name TEXT,
SNP_Index INTEGER,
SNP_Aux INTEGER,
Sample_ID TEXT,
SNP TEXT,
Allele1_Top TEXT, 
Allele2_Top TEXT, 
Allele1_Forward TEXT, 
Allele2_Forward TEXT, 
Allele1_AB TEXT, 
Allele2_AB TEXT, 
Chr TEXT, 
Position TEXT, 
GC_Score INTEGER,
GT_Score INTEGER,
Theta INTEGER,
R INTEGER,
B_Allele_Freq INTEGER,
Log_R_Ratio INTEGER)''')
```

```python
genstudio.columns = ['SNP_Name', 'SNP_Index', 'SNP_Aux', 'Sample_ID', 'SNP', 'Allele1_Top', 'Allele2_Top', 
                     'Allele1_Forward', 'Allele2_Forward', 'Allele1_AB', 'Allele2_AB', 'Chr', 'Position', 
                     'GC_Score', 'GT_Score', 'Theta', 'R', 'B_Allele_Freq', 'Log_R_Ratio'] 
```

```python
genstudio.to_sql('genstudio', con=connection, if_exists='append', index = False)
```

```python
connection.execute('''CREATE TABLE metadata (
sample_id INTEGER PRIMARY KEY,
dna_chip_id TEXT,
breed TEXT,
sex TEXT)''')
```

```python
metadata.to_sql('metadata', con=connection, if_exists='append', index = False)
```

```python
connection.commit()
```

```python
connection.execute('''CREATE TABLE IF NOT EXISTS link_table_cascade (
attendance_id INTEGER PRIMARY KEY,
sample_id INTEGER,
SNP_id INTEGER,
FOREIGN KEY (sample_id) REFERENCES metadata (sample_id) ON DELETE CASCADE,
FOREIGN KEY (SNP_id) REFERENCES learners (SNP_id) ON DELETE CASCADE
)''')
```

```python
link_table.to_sql('link_table_cascade', con=connection, if_exists='append', index = False)
```
